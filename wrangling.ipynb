```python
import pandas as pd
```

```python
wkly_data = pd.read_csv("7-day-input-window.csv")
biwkly_data = pd.read_csv("14-day-input-window.csv")
mnthly_data = pd.read_csv("30-day-input-window.csv")
```

```python
for df in (wkly_data, biwkly_data, mnthly_data):
    df.dropna(inplace=True)
    df.reset_index(drop=True, inplace=True)
```

```python
s = 30
data = mnthly_data
direction = []
monthly_returns = []
for i in range(len(data)):
    try:
        monthly_return = (data.Close.loc[i+s] - data.Close.loc[i]) / data.Close.loc[i]
        monthly_returns.append(monthly_return)
        if monthly_return > 0:
            direction.append("Up")
        else:
            direction.append("Down")
    except:
        break
```

```python
len(direction)
```

```python
import numpy as np
direction[len(direction):len(mnthly_data)+1] = [np.nan]*(len(mnthly_data)-len(direction))
monthly_returns[len(monthly_returns):len(mnthly_data)+1] = [np.nan]*(len(mnthly_data)-len(monthly_returns))
print(len(direction))
print(len(monthly_returns))
```

```python
mnthly_data["Direction"] = direction
mnthly_data["30DayReturn"] = monthly_returns
```
