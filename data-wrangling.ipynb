```python
import pandas as pd
import yfinance as yf
import talib
```

```python
tickers = sorted(list(set(pd.read_csv("C:/Users/Robpr/OneDrive/Documents/Projects/USDistressPrediction/us-income-annual.csv", sep=";").Ticker)))
```

14 day, 30 day, 3 months, 6 months, 12 months
- Simple Moving Average (SMA): 
$SMA_n = 1/n \sum_{i=0}^{n-1} C_{t-i}$
- Exponential Moving Average (EMA): $EMA_n = \sum_{i=0}^{n-1}w_iC_{t-i}$
where $\sum_{i=0}^{n-1}w_i = 1$ and $n$ is the input window length.
- Average True Range (ATR): $ATR_n = EMA_n(\max\{H_t - L_t,\ |H_t-C_{t-1}|,\ |L_t-C_{t-1}|\})$
where $H_t$, $L_t$ and $C_t$ are the high, low and closing prices on a day $t$, respectively.
- Average Directional Movement Index (ADMI):



```python
historical_data = []
available_tickers = []
for ticker in tickers:
    try:
        tick = yf.Ticker(ticker)
        hist = tick.history(start="2020-01-01", end="2020-12-31")
        hist["Ticker"] = ticker        
        historical_data.append(hist)
        available_tickers.append(ticker)
    except:
        pass
historical_data[:5]
```

```python
# pd.Series(available_tickers).to_csv("available-yfinance-tickers.csv")
```

```python
for hist_data in historical_data:
    try:
        # Overlap Studies
        hist_data["SMA"] = talib.SMA(hist_data.Close, timeperiod=30)
        hist_data["EMA"] = talib.EMA(hist_data.Close, timeperiod=30)
        hist_data["WMA"] = talib.WMA(hist_data.Close, timeperiod=30)
        hist_data["TEMA"] = talib.TEMA(hist_data.Close, timeperiod=30)

        # Momentum Indicators
        hist_data["ADMI"] = talib.ADX(high=hist_data.High, low=hist_data.Low, close=hist_data.Close, timeperiod=30)
        hist_data["CCI"] = talib.CCI(high=hist_data.High, low=hist_data.Low, close=hist_data.Close, timeperiod=30)
        hist_data["ROC"] = talib.ROC(hist_data.Close, timeperiod=30)
        hist_data["RSI"] = talib.RSI(hist_data.Close, timeperiod=30)
        hist_data["Williams"] = talib.WILLR(high=hist_data.High, low=hist_data.Low, close=hist_data.Close, timeperiod=30)
        stochfk, stochfd = talib.STOCHF(high=hist_data.High, low=hist_data.Low, close=hist_data.Close, fastk_period=30, fastd_period=30, fastd_matype=0.7)
        hist_data["STOCHFK"] = stochfk
        hist_data["STOCHFD"] = stochfd
        
        # Volume Indicators
        hist_data["ChaikinAD"] = talib.AD(hist_data.High, hist_data.Low, hist_data.Close, hist_data.Volume)
        
        hist_data.reset_index(level="Date")
    except:
        pass
```

```python
data = pd.concat(historical_data)
```
